# 파일을 불러와서 그래프로 시각화

```python
# 라이브러리 생성
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# 그래프에 표시할 파일 불러오기
groupDataS01 = pd.read_csv('./st0001data.csv')
groupDataS02 = pd.read_csv('./st0002data.csv')

# x축 y축 지정
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME


plt.xticks( range(min(x),max(x),10) ) # 숫자(10)는 x축 범위 지정
plt.xlabel('week') # x축 제목
plt.ylabel('avg_volume') # y축 제목
plt.title('weekly trend') # 차트 이름
plt.plot(x,y)
```

# 그래프 하나에 또 다른 데이터를 넣어 비교해보기

```python
# 라이브러리 생성
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# 그래프에 표시할 파일 불러오기
groupDataS01 = pd.read_csv('./st0001data.csv')
groupDataS02 = pd.read_csv('./st0002data.csv')

# x,y,y2축 지정 (y2 = 다른 파일의 특정 데이터와 비교하기 위함)
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME



plt.xlabel('week') # x축 제목
plt.ylabel('avg_volume') # y축 제목
plt.title('weekly trend') # 차트 이름
plt.grid(True) # 그리드 표시 (차트 안의 선)
plt.plot(x,y, label='ST0001') # 라벨
plt.plot(x,y2, label='ST0002') # 라벨
plt.legend(loc=0) # 라벨의 위치 선정
```

# 불러올 파일에서 특정 데이터만 추출해서 시각화

```python
# 라이브러리 생성
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# 그래프에 표시할 파일 불러오기
featuresData2015 = pd.read_csv('../dataset/feature_regression_example.csv')
# 파일에서 특정 데이터만 추출 (year컬럼에서 2015년도의 데이터만)
featuresData2015 = featuresData2015.loc[ featuresData2015.YEAR == 2015 ]

# x,y축 설정
x=featuresData2015.WEEK
y=featuresData2015.QTY


# bo = b = 블루, o = 데이터 이음새의 형태 
plt.plot(x,y, 'bo-')
plt.xlabel('WEEK') # x축 제목
plt.ylabel('QTY') # y축 제목

#plt.title('weekly trend') # 차트 이름
#plt.grid(True) # 그리드 표시 (차트 안의 선)
#plt.plot(x,y, label='ST0001')
#plt.plot(x,y2, label='ST0002')
#plt.legend(loc=0)
```

# 2개의 그래프 만들기

```python
# 라이브러리 생성
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


# 그래프에 표시할 파일 불러오기
groupDataS01 = pd.read_csv('./st0001data.csv')
groupDataS02 = pd.read_csv('./st0002data.csv')

# x,y,y2축 설정
x=groupDataS01.WEEK
y=groupDataS01.MEAN_VOLUME
y2=groupDataS02.MEAN_VOLUME

# 그래프의 크기 설정
fig = plt.figure(figsize=(14,4))

# 121의 의미는 1행 2열 중 첫번째에 표시하겠다는 뜻
ax1 = fig.add_subplot(121)
# g = 그린, o = 이음새 표시형태
ax1.plot(x,y, 'go-')
# x,y축 제목
plt.xlabel('week') 
plt.ylabel('avg_volume') 

# 121의 의미는 1행 2열 중 두번째에 표시하겠다는 뜻
ax2 = fig.add_subplot(122)
# b = 블루, o = 이음새 표시형태
ax2.plot(x,y2, 'bo-')
# x,y축 제목
plt.xlabel('Week')
plt.ylabel('avg_volume')
```
